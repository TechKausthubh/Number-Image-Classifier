```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
print('Number of training samples:', train_images.shape[0])
print('Number of test samples:', test_images.shape[0])
print(f"Shape of the images in the training dataset: {train_images[0].shape}")
```

```python
fig, axes = plt.subplots(1, 10, figsize=(10, 10))
for i in range(10):
    axes[i].imshow(train_images[i].reshape(28, 28), cmap='gray')
    axes[i].set_title(f"Label: {train_labels[i]}")
    axes[i].axis('off')
plt.show()
```

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {test_acc:.2f}')
```

```python
fig, axes = plt.subplots(1, 10, figsize=(10, 10))
for i in range(10):
    axes[i].imshow(test_images[i].reshape(28, 28), cmap='gray')
    axes[i].set_title(f"Label: {test_labels[i]}")
    axes[i].axis('off')
plt.show()
```

```python
# Preditions on first 10 test images
for i in range(10):
    probabilities = model.predict(test_images[i].reshape(1, 28, 28, 1))
    # print(probabilities)
    predicted_label = np.argmax(probabilities)
    print(f'Predicted label: {predicted_label}, True label: {test_labels[i]}')
```
